## Abundance evolution plots
This uses the plot method.

```python
specs=['PROT','HE  4','C  12','N  14','O  16']
i=0
for spec in specs:
    px.plot('time',spec,logy=True,logx=True,shape=utils.linestyle(i)[0],show=False,title='')
    i += 1
ylim(-5,0.2)
legend(loc=0)
xlabel('$\log t / \mathrm{min}$')
ylabel('$\log X \mathrm{[mass fraction]}$')
```
