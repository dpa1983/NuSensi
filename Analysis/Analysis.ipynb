#####Abstract

To calculate sensitivity factors, e.g. $S_{ij}$ for isotope $i$ and rate $j$
in form of a matrix the script sensitivity.py is used in the following. For the analysis the results from the rate modifications and a reference
run (without any modifications) is needed. The results
in the path_results dir (after performing runs, see nusensi.input)
can be easily used. Additionaly one can calculate the errors
of each sensitivity factor by taking into account the errors from
rates specified in errors.txt. Furthermore plot_sens.py is used to visualize
the sensitivity with a isotopic chart.

```python
#import modules
import sensitivity as sens
import plot_sens as ps
%pylab nbagg
```

### Read run with default rates

```python
runs=sens.read_ref_run(cycle=125,path='../run_r45f1.0')
```

### Create the sensitivity matrix

assuming that all results in ../results needs to be 
processed and the rate factor is 0.9:


```python
runs.sens_matrix(path_results=["../results/"],factors=[1.1],
                 filename='sensitivity_matrix.txt',decay=False,abu_threshold=1e-9)
```

Voila! Now have a look at the file 'sensitivity_matrix.txt'. See the docstring of the function calc_matrix for a explanation about the structure.

## Plot sensitivities

### Read in matrix

```python
m1=ps.plot_sens('sensitivity_matrix.txt')
```

### Create chart plot

Now you can calculate the errors of each sensitivity factor by creating a new matrix. Note for that you need
to provide the errors of all rates in the sensitivity matrix in errors.txt and a list of isotopes stable.txt
In case the provided list is sufficient use:


```python
m1.abu_chart(ifig=111,sens_param_type='reaction',sens_param='C  13(a,n)')
```
